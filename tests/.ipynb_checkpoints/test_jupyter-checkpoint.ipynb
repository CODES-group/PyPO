### NOTEBOK EXAMPLE IN SPYDER - INVERSE MAPPING OF SHOWER PROBLEM 

```python
import numpy as np
import time
import nbconvert

import sys
sys.path.append('../')
from  src.pyprop import nlp_based_approach
```

```python
def shower2x2(u):
    
    d = np.zeros(2)
    y = np.zeros(2)
    y[0]=u[0]+u[1]
    if y[0]!=0:
        y[1]=(u[0]*(60+d[0])+u[1]*(120+d[1]))/(u[0]+u[1])
    else:
        y[1]=(60+120)/2
        
    return y
```

```python
u0 = np.array([0, 10])
lb = np.array([0, 0])
ub = np.array([100,100])

DOS_bound = np.array([[17.5, 21.0],
                    [80.0, 100.0]])

DOSresolution = [10, 10]
    

t = time.time()
fDIS, fDOS, message = nlp_based_approach(DOS_bound, 
                                          DOSresolution, 
                                          shower2x2, 
                                          u0, 
                                          lb,
                                          ub, 
                                          method='ipopt', 
                                          plot=True, 
                                          ad=False)
    
    
elapsed = time.time() - t
```

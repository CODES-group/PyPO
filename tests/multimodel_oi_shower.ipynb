```python
import sys
sys.path.append('../')

from pyprop import multimodel_rep, OI_calc

import numpy as np
```

```python
def shower2x2(u):
    
    d = np.zeros(2)
    y = np.zeros(2)
    y[0]=u[0]+u[1]
    if y[0]!=0:
        y[1]=(u[0]*(60+d[0])+u[1]*(120+d[1]))/(u[0]+u[1])
    else:
        y[1]=(60+120)/2
        
    return y
```

Defining The AIS and DOS Bounds

```python
DOS_bounds =  np.array([[10, 20], 
                        [70, 100]])

AIS_bounds =  np.array([[0.1, 10],
                        [0.1, 10]])

AIS_resolution =  [5, 5]

model =  shower2x2
```

Obtaining the AOS from the AIS

```python
AOS_region  =  multimodel_rep(AIS_bounds, 
                AIS_resolution, model, polytopic_trace = 'polyhedra')
```

```python
OI = OI_calc(AOS_region,
            DOS_bounds, hypervol_calc= 'robust')
```
